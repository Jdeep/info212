#### Qestion 3:
Given the following list of file names: 
```
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']
```
Write a program to plot the daily averages, maximums, and minimums of inflammation for each of the file. Your plot should have three rows each of which shows side-by-side plots of average, max, and min. The following figure illustrates the expected outcome:
![](assignment2-question3.png)

````python
def Inflam_plot(path_list):
    for path in path_list:
        data = np.loadtxt('datasets/inflammation-data/'+path, delimiter = ',')
        #get avg, max, and min
        daily_avg = np.mean(data, axis = 0)
        daily_max = np.max(data, axis = 0)
        daily_min = np.min(data, axis = 0)
        #plot data
        fig = plt.figure(figsize=(12,3))
        axis1 = fig.add_subplot(131)
        axis2 = fig.add_subplot(132)
        axis3 = fig.add_subplot(133)
        #avg plot
        axis1.plot(daily_avg)
        axis1.set_title('Average Inflammation in Patients by Day')
        axis1.set_xlabel('Average')
        axis1.set_ylabel('Patients')
        #max plot
        axis2.plot(daily_max)
        axis2.set_title('Maximum Inflammation in Patients by Day')
        axis2.set_xlabel('Average')
        axis2.set_ylabel('Patients')
        #min plot
        axis3.plot(daily_min)
        axis3.set_title('Minimum Inflammation in Patients by Day')
        axis3.set_xlabel('Average')
        axis3.set_ylabel('Patients')

        plt.tight_layout()
#Get file data
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']
Inflam_plot(filenames)
````
